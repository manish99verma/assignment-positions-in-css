2. Explain the difference between Absolute and Relative positioning.
- Difference between Absolute and Relative Postioning.
  
 Relative Positioning:
  *   It moves the element relative to its normal position.
  *   The space it originally occupied still remains.
  *   Used when you want to nudge an element slightly from where it normally
      sits.

  Example:
  ```
  box {
  position: relative;
  top: 10px;
  left: 20px;
  }
  ```

  Absolute Positioning  
  *   It removes the element from the normal flow.
  *   The element is placed relative to the nearest positioned ancestor (if no positioned parent, then relative to `<body>`).
  *  It doesn't leave any space where it used to be.

  Example:
  ```
  box {
  position: absolute;
  top: 10px;
  left: 20px;
  }
  ```

5. Explain the z-index, with a code example.
-  Z- index: z-index is a CSS property that helps us decide which element should be on top when two or more elements are overlapping. You can think of it like layers — the one with higher z-index will be on top. But for z-index to work, the element must have a position like relative, absolute, fixed, or sticky.

##Code Example
In this code, there are two boxes.
- The red one has z-index: 1
- The blue one has z-index: 2

So when they overlap, the blue box will appear on top of the red one.

```Code
<!DOCTYPE html>
<html>
  <head>
    <style>
      .box1 {
        width: 200px;
        height: 200px;
        background-color: red;
        position: absolute;
        top: 50px;
        left: 50px;
        z-index: 1;
      }

      .box2 {
        width: 200px;
        height: 200px;
        background-color: blue;
        position: absolute;
        top: 100px;
        left: 100px;
        z-index: 2;
      }
    </style>
  </head>
  <body>    
    <div class="box1"></div>
    <div class="box2"></div>
  </body>
</html>

```
